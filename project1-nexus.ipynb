```python
import random

# Greeting function
def greet_user():
    print("Hi there! I'm your friendly chatbot, Chad.")

# Farewell message
def farewell():
    print("It was a pleasure chatting with you! Until next time.")

# Function to handle chatbot responses
def respond(user_input):
    # Define a dictionary of responses for different questions
    responses = {
        "what can you do?": [
            "I can answer your questions on various topics, tell you jokes, and even have simple conversations!",
            "I'm still under development, but I'm learning new things every day."
        ],
        "what is your purpose?": [
            "My purpose is to assist you with information and provide friendly conversation.",
            "I'm here to make your day a little brighter with some knowledge and humor!"
        ],
        "favorite color": ["I don't have emotions like humans, but I find shades of blue calming and serene.", 
                           "What's your favorite color? Tell me about it!"],
        "tell me a joke":["Sure, here's a joke for you:  Why don't scientists trust atoms?    Because they make up everything!",
                         "Sure, here's a joke for you:  Why did the scarecrow win an award?   Because he was outstanding in his field!"],
    
    }

    # Check for matching question or keywords in user input
    for key, value in responses.items():
        if key in user_input.lower() or any(keyword in user_input.lower() for keyword in key.split(" ")):
            response = random.choice(value)
            print(response)
            break

    # Handle unrecognised questions
    else:
        print("I'm not quite sure what you mean. Could you please rephrase your question?")

# Conversation flow with personalization
def start_chat():
    greet_user()

    # Remember user's name
    name = input("What's your name? ")
    print(f"Nice to meet you, {name}! How can I help you today?")

    # Ask initial questions and react to answers
    for question in ["What's your favorite hobby?", "Do you have any plans for the weekend?", "Tell me something interesting about yourself"]:
        print(question)
        user_input = input(">> ")
        # Personalize responses based on user input
        if "hobby" in question and "reading" in user_input:
            print(f"That's great, {name}! I love reading too.")
        elif "weekend" in question and "relaxing" in user_input:
            print(f"Sounds like a good plan, {name}. Enjoy your well-deserved rest!")
        else:
            print(f"That's fascinating, {name}! Thanks for sharing.")

    while True:
        user_input = input(">> ")
        if user_input.lower() == "bye":
            farewell()
            break
        else:
            respond(user_input)

if __name__ == "__main__":
    start_chat()
```
